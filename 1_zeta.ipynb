```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
data = pd.read_csv("1_zeta_results.csv")

# Graficar 
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Subplot izquierdo
plt.plot(data["k"], data["Zeta(4)"], "b-", label="Aproximación ζ(4)")
plt.axhline(y=data["Exact_Value"].iloc[0], color="r", linestyle="--", label="Valor exacto (π⁴/90)")
plt.xlabel("Número de términos (k)")
plt.ylabel("ζ(4)")
plt.title("Aproximación de ζ(4) vs k")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
# Graficar Error vs k 
plt.subplot(1, 2, 2)  # Subplot derecho
plt.plot(data["k"], data["Error"], "g-", label="Error absoluto")
plt.xlabel("Número de términos (k)")
plt.ylabel("Error")
plt.title("Error de la aproximación vs k")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.savefig("zeta_vs_k_results.png", dpi=300)
plt.show()
```
